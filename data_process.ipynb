```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv(u'C:\\Users\Administrator\Downloads\data.csv',encoding='gbk')
```

```python
data.drop(['custid', 'trade_no', 'bank_card_no', 'id_name'], axis=1, inplace=True)## 删除与个人身份相关的列
date_feature=['first_transaction_time', 'latest_query_time', 'loans_latest_time']#提取时间特征
data['first_transaction_time'].fillna(data['first_transaction_time'].median(), inplace=True)  #有缺失值，先填充，众数填充
data['first_transaction_time'] = data['first_transaction_time'].apply(lambda x:str(x)[:4] + '-' + str(x)[4:6] + '-' + str(x)[6:8])#数据类型转为字符串类型
```

```python
data['first_transaction_time']=pd.to_datetime(data['first_transaction_time'])
#新增时间序列特征，具有时间戳的特征分别衍生 年，月，星期。
data.loc[:,'first_transaction_time_year']=data['first_transaction_time'].apply(lambda x : x.year)
data.loc[:,'first_transaction_time_month']=data['first_transaction_time'].apply(lambda x : x.month)
data.loc[:,'first_transaction_time_weekday']=data['first_transaction_time'].apply(lambda x : x.dayofweek)

data['latest_query_time']=pd.to_datetime(data['latest_query_time'])#字符串转换成时间戳
data.loc[:,'latest_query_time_year']=data['latest_query_time'].apply(lambda x : x.year)
data.loc[:,'latest_query_time_month']=data['latest_query_time'].apply(lambda x : x.month)
data.loc[:,'latest_query_time_weekday']=data['latest_query_time'].apply(lambda x : x.dayofweek)

data['loans_latest_time']=pd.to_datetime(data['loans_latest_time'])
data.loc[:,'loans_latest_time_year']=data['loans_latest_time'].apply(lambda x : x.year)
data.loc[:,'loans_latest_time_month']=data['loans_latest_time'].apply(lambda x : x.month)
data.loc[:,'loans_latest_time_weekday']=data['loans_latest_time'].apply(lambda x : x.dayofweek)
```

```python
# 填充缺失值
data['latest_query_time_year'].fillna(data['latest_query_time_year'].median(), inplace = True)
data['latest_query_time_month'].fillna(data['latest_query_time_month'].median(), inplace = True)
data['latest_query_time_weekday'].fillna(data['latest_query_time_weekday'].median(), inplace = True)

data['loans_latest_time_year'].fillna(data['loans_latest_time_year'].median(), inplace = True)
data['loans_latest_time_month'].fillna(data['loans_latest_time_month'].median(), inplace = True)
data['loans_latest_time_weekday'].fillna(data['loans_latest_time_weekday'].median(), inplace = True)

data.drop(date_feature,axis=1,inplace=True)#删除之前的特征
data.drop(columns='Unnamed: 0',axis=1,inplace=True)#有一列无列名。类似于id 可删除

print (data['student_feature'].isnull().sum())
data['student_feature'].fillna(0,inplace=True)# student_feature 缺失严重达一半以上， 用0填充
print(data['student_feature'].value_counts())#学生特征状态0,1,2 三个状态

print(data['reg_preference_for_trad'].value_counts())#城市的类别有5个，全是非字符串，准备将其用数字代替，方便后面训练

dict_city=dict()
dict_city={'一线城市':0,'二线城市':1,'境外':2,'三线城市':3,'其他城市':4}
```

```python
data['reg_preference_for_trad']=data['reg_preference_for_trad'].map(dict_city)
print (data['reg_preference_for_trad'])
```

```python
for col in data.columns:
    summ = data[col].isnull().sum()
    if summ:
        data[col].fillna(data[col].median(), inplace = True)
        
data
```
